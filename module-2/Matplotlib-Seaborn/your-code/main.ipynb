#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4

* Explore plt.xticks

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
import calendar

plt.bar(fitbit['Months_encoded'].unique(),
         fitbit['Steps'].groupby(fitbit['Months_encoded']).median())
plt.xticks(np.arange(12), calendar.month_name[1:13], rotation=20)
plt.show()


# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
plt.bar(fitbit['Work_or_Weekend'].unique(),
         fitbit['Steps'].groupby(fitbit['Work_or_Weekend']).median())
plt.xticks(np.arange(2))
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
variables = ['Minutes Lightly Active','Minutes Very Active', 'Minutes Sedentary']
for x in variables:
    plt.scatter(fitbit['Steps'], fitbit[x])
```

```python
variables = ['Minutes Lightly Active','Minutes Very Active','Minutes Sedentary']
for x in variables:
    figure = plt.figure
    ax = plt.gca()
    ax.scatter(fitbit['Steps'], fitbit[x])
    ax.set_xlabel('Steps')
    ax.set_ylabel(x)
    ax.set_title("{} vs {}".format('Steps', x))
    plt.show()
```
